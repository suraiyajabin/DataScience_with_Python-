```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.linspace(-10,10,60) #generate 30 equispaced points between -10 to 10
```

```python
a
```

```python
b=a**4
```

```python
b
```

```python
b1=np.tanh(a)
b2=np.cos(a)
b3=np.sin(a)
```

```python
b1
```

```python
x=np.arange(1,10)
y=2*x
```

```python
x
```

```python
y
```

```python
#using plot function
plt.plot(x,y)
```

```python
#using plot function
plt.plot(a,b1)
```

```python
#using title, xlabel, ylabel
#from IPython.display import display  # Import display function
plt.plot(a,b1)
plt.title("Plot of Tanh function")
plt.xlabel("X values")
plt.ylabel("Tanh values")
#display(plt.gcf())  # Force display the current figure
plt.savefig("tanh_plot.png")  # Save as an image
```

```python
#using Figure object
fig=plt.Figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
plt.show()
```

```python
fig
```

```python
#Now using OOP based Figure() object; here with one figure you can have multiple axes of different sizes
fig2=plt.Figure(figsize=(3,3),dpi=100) 
#large axes
axes1=fig2.add_axes([0,0,1,1])
axes1.plot(a,b)
#small axes
axes2=fig2.add_axes([0.2,0.5,0.3,0.3])
axes2.plot(a,b)
fig2.savefig("2axes_plot.png",bbox_inches='tight')
plt.show()
```

```python
fig2
```

```python
#using subplots 1 X 3
fig4,axes4=plt.subplots(nrows=1,ncols=3,figsize=(12,4))
axes4[0].plot(x,y)
axes4[1].plot(a,b)
axes4[2].plot(a,b1)
```

```python
#using subplots 2 X 2
fig5,axes5=plt.subplots(nrows=2,ncols=2,figsize=(6,6))
axes5[0][0].plot(x,y)
axes5[0][0].set_ylabel('y label')
axes5[0][0].set_xlabel('x label')
axes5[0][0].set_title('Straight Line Plot')
axes5[0][0].set_xlim(2,6)
axes5[0][1].plot(a,b)
axes5[1][0].plot(a,b1)
axes5[1][1].plot(a,b2)
plt.tight_layout()
fig5.subplots_adjust(hspace=0.3,wspace=0.2)
fig5.suptitle('Figure level title',fontsize=18, horizontalalignment='left')
```

```python
#drawing multiple plots in one Graph
fig6,axes6=plt.subplots()
#axes6=fig6.add_axes([-10,-10,10,10])
#possible linestyle option can take values '+','*' '--','-.','steps'
axes6.plot(a,b3,label='sin',color='red',lw=0.5,ls='--') 
# alternatively you can select color code using "Hex Color picker"
#check matplotlib markers api on google to get full list of markers https://matplotlib.org/stable/api/markers_api.html
axes6.plot(a,b1,label='tanh',color='#6e6b12',lw=0.5,linestyle='-.',marker='^',ms=5.5) 
#just google "Hex color picker", and select code for your customised color
axes6.plot(a,b2,label='cos', color='#7ecf21',lw=2,linestyle=':',marker='+',ms=8) 
#try option of marker and markersize/ms from https://matplotlib.org/stable/api/markers_api.html
#lw option is used to set linewidth, default=1, if you set it 5, it's now 5 times lw=1
axes6.legend(loc=0)
#ax6.legend(loc=(1.5,0.5))
```
