```python
j=np.linspace(0,1,400)
color1=np.random.randint(0, 4, 400) #generates 400 random int numbers lying bw 0 to 3, 4 is exclusive
k=np.arange(0,400)
r=np.random.normal(120,10,400) #generate data around value 120, with sd=10, how many points=400
plt.scatter(k,r,s=6,edgecolors='face',c=color1,cmap='hsv',marker='D') 
# try markers = ['o', 's', '^', 'D']  # Circle, Square, Triangle, Diamond
```

```python
r
```

Why Do Some Points Have the Same Color?
Since there are 400 points but only 12  with cmap='set10' option, Matplotlib reuses colors.
For example:
k=0  Color 1
k=1  Color 2
...
k=11  Color 12
k=12  Color 1 (repeats)
k=13  Color 2 (repeats)
and so on...
So, every 12th point gets the same color.

```python
plt.scatter(k,r,s=3,edgecolors='face',c=color1,cmap='viridis')
```
