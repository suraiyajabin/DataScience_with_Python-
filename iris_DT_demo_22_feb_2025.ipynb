```python
import numpy as np
import pandas as pd
```

```python

df=pd.read_csv("iris.data",header=None,names=['sepal_len','sepal_width','petal_len','petal_width','class']) 
 #if your csv doesn't have a header, header=0, if first row has column headers
 #use this for csv with header df=pd.read_csv("test.csv",header=0)
 #to assign column names while reading a csv using pandas with no column header in it use names=['','',....]
```

```python
df
```

```python
from sklearn.preprocessing import Normalizer, normalize
```

```python
X=df.iloc[:,0:4]  #select just first 4 columns with range 0:4, 4 excluded
```

```python
X.head()
```

```python
max_normalizer_row=Normalizer(norm='max') #it can take l1, l2, max row-wise
#Decision Tree/Random Forest models do not require pre-processed data.
#including normalization here to keep up ML implementation of pipelines.
```

```python
X_norm_1_row=max_normalizer_row.fit_transform(X)  # it calculates row wise Max norm which is not needed
```

```python
X_norm_1_row
```

```python
max_normalizer2=Normalizer(norm='max') #it can take l1, l2, max
 
# Apply normalization column-wise
X_trans=X.transpose() #if you don't transpose, it will again do row wise normalization
l_max_norm_col = max_normalizer2.fit(X_trans)
l_max_norm_col=max_normalizer2.transform(X_trans)
#we can invoke fit and transform separately and together too using fit_transform(X_trans).
```

```python
l_max_norm_col.shape
```

```python
X_col_norm=l_max_norm_col.transpose()
```

```python
X_col_norm.shape
```

```python
X_col_norm
```

```python
#for inline Normalization using normalize method
# Apply column-wise normalizations
from sklearn.preprocessing import normalize
X_norm1 = normalize(X, norm='max', axis=0)  #axis=0 option suggests normalize data column wise with l1, l2, max
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.normalize.html
```

```python
X_norm1
```

```python
Y=df.iloc[:,4]
```

```python
Y
```

```python
# Split data into training and testing sets (80% train, 20% test)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_norm1, Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Create and train the Decision Tree classifier
model_DT = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
model_DT.fit(X_train, y_train)  # Train the model
'''' Hyperparameters to Tune
criterion='gini' or 'entropy' â†’ Measure for splitting nodes.
max_depth=3 â†’ Limits tree depth to prevent overfitting.
min_samples_split=2 â†’ Minimum samples required to split a node.
min_samples_leaf=1 â†’ Minimum samples per leaf node.'''
```

```python
# Predict on test data
y_pred = model_DT.predict(X_test)
```

```python
# Evaluate accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    precision_recall_curve,
    roc_curve,
    auc
)
# Display classification report
print("\nClassification Report:")
#print(classification_report(y_test, y_pred, target_names=df.target_names))
print(classification_report(y_test, y_pred))
```

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
Y_labels=['Iris-setosa','Iris-versicolor','Iris-virginica']
```

```python
# Plot the confusion matrix
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=Y_labels, yticklabels=Y_labels)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
#Visualize the Decision Tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plot_tree(model_DT, feature_names=['sepal_len','sepal_width','petal_len','petal_width','class'] , class_names=['Iris-setosa','Iris-versicolor','Iris-virginica'], filled=True)
plt.show()
```
