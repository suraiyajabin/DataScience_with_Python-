Demo of alternate method to OHS brand column
we will OHE only those categorical values of brand which have value_count()>threshold
remaining will be given a new categorical value 'other'
which means anything other than those 9 values  > threshold

```python
df['brand'].value_counts()
```

```python
counts=df['brand'].value_counts()
#value_counts() method in pandas is used to count the occurrences of unique values in a Series.
```

```python
counts.shape
```

```python
df['brand'].nunique()
```

```python
threshold=200
rep=counts[counts<=threshold].index
#conditional selection using DataFrame
'''we will OHE only those categorical values of brand which have value_count()>threshold
remaining will be given a new categorical value 'other'
which means anything other than those 9 values  
counts <= threshold creates a Boolean mask, marking values that occur ≤ threshold times.
Applying this mask (counts[counts <= threshold]) keeps only rows where 
the count is below or equal to threshold.
.index gets the index values of the filtered Series.'''
```

```python
rep.shape
```

```python
pd.get_dummies(df['brand'].replace(rep,'other'))
```

```python
df_brand=pd.get_dummies(df['brand'].replace(rep,'others'),drop_first=True,dtype=np.int32)
#we are separately handling OHE of brand column for all rows/samples
```
