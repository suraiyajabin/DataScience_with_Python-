```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(6,6),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(5,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education')
```

```python
#use hue=another categorical column name to give a third dimension to your scatter plot
plt.figure(figsize=(5,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,hue='division')
```

```python
plt.figure(figsize=(4,4),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,hue='salary')
```

```python
#choosing different colors for legends and dots in scatter plot
#https://matplotlib.org/stable/users/explain/colors/colormaps.html
plt.figure(figsize=(5,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,hue='division',palette='tab10') #palette='Pastel1'/'Pastel2'/'Dark2'/'tab10'
```

```python
#choosing different colors for legends and dots in scatter plot
#https://matplotlib.org/stable/users/explain/colors/colormaps.html
plt.figure(figsize=(5,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,hue='division',palette='Dark2') #palette='Pastel1'/'Pastel2'/'Dark2'/'tab10'
#Ignoring `palette` because no `hue` variable has been assigned as they work together
```

```python
#using size option to show small points for small values, and so on
plt.figure(figsize=(5,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,size='training level')
```

```python
#using s anf alpha parameters to control size (s) and transparency of points (alpha)
plt.figure(figsize=(5,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,s=15,alpha=0.6)
```

```python
#using style to get different markers for different categorical values of a column

plt.figure(figsize=(10,10),dpi=150)
sns.scatterplot(x='salary',y='sales',data=df,style='training level')
```

```python
#combining style to get different markers for different categorical values of a column
#with palette and hue

plt.figure(figsize=(10,10),dpi=150)
sns.scatterplot(x='salary',y='sales',data=df,style='training level',hue='training level',palette='Dark2')
plt.savefig("scatter_plot_sales_sal.png")
```
